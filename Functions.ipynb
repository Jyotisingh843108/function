# 1. What is the difference between a function and a method in Python?

  - A function is a block of reusable code that performs a specific task. It is defined using the def keyword and is written outside of any class. Functions are independent and can be called directly by using their name, followed by parentheses, and passing any required arguments. They do not automatically take any object or instance as a parameter. For example, if we define a function like def greet(name): return "Hello " + name, we can simply call it as greet("Jyoti"). Functions are commonly used when we want to perform operations that are not tied to any specific object or class, making them useful for general-purpose tasks.

  -A method, on the other hand, is a function that is defined inside a class and is associated with the objects created from that class. The first parameter of a method is always self, which refers to the instance of the class. This allows the method to access or modify the attributes of the object it belongs to. Methods are called using the object name, like object_name.method_name(). For example, if we define a method def greet(self, name) inside a class called Person, we must first create an object such as p = Person() and then call the method as p.greet("Jyoti"). So, while both functions and methods perform tasks, methods are specifically designed to work within the context of objects and classes.

# 2. Explain the concept of function arguments and parameters in Python.

  - parameters are the variable names used in a function definition. They act as placeholders for the values that will be passed to the function. For example, in def greet(name):, the word name is a parameter.

  - Arguments are the actual values we give to the function when we call it. For example, if we call the function like greet("Jyoti"), the value "Jyoti" is the argument. So, parameters are used when defining a function, and arguments are used when calling it.

# 3. What are the different ways to define and call a function in Python?

  - we can define a function using the def keyword, followed by the function name and parentheses. Functions can be defined in several ways: without parameters (def hello():), with parameters (def greet(name):), with default values (def greet(name="Guest"):), or with variable arguments using *args and **kwargs. We can also define anonymous functions using lambda, like lambda x: x * 2.

  - To call a function, we write the function name followed by parentheses. If the function requires parameters, we pass the values (arguments) inside the parentheses. For example, if we have def add(a, b):, we call it as add(2, 3). We can also use keyword arguments when calling, like add(a=2, b=3).

# 4.  What is the purpose of the `return` statement in a Python function?

  - The purpose of the return statement in a Python function is to send a result back to the part of the program that called the function. When a function reaches a return statement, it stops executing and passes the specified value back to the caller. This allows the function to produce an output that can be stored, used in calculations, or printed. Without a return statement, a function returns None by default.

  example:

  def add(a, b):

    sum = a + b

    return sum

  result = add(3, 5)

 print(result)  # Output: 8

 The function add takes two numbers, adds them, and uses return to send the sum back. When we call add(3, 5), it returns 8, which we store in result and then print.





# 5. What are iterators in Python and how do they differ from iterables?

  - An iterable is any object that we can loop over, such as a list, string, or dictionary. It provides a way to get an iterator, which is an object that lets we access its elements one at a time. An iterator keeps track of the current position during the iteration and has a __next__() method to move to the next element. When there are no more elements, the iterator raises a StopIteration exception.

  - For example, a list [10, 20, 30] is an iterable. When you create an iterator from this list using the iter() function, you get an iterator object. Using next() on this iterator will return 10 first, then 20, then 30. After that, calling next() again will raise StopIteration because the iterator has reached the end.



# 6. Explain the concept of generators in Python and how they are defined.

  - Generators in Python are special functions that produce items one at a time and only when requested, which makes them memory-efficient. Unlike regular functions that use return to send back a value and end, generators use the yield keyword to provide a value and pause the function, resuming from where they left off when called again. This allows you to handle large sequences or streams of data without storing everything in memory.

  -For example, here is a generator function that yields numbers from 1 to 3:

  def my_generator():

    yield 1

    yield 2

    yield 3

  - When you call my_generator(), it returns a generator object. You can retrieve each value by calling next() repeatedly, getting 1, then 2, then 3, until it raises StopIteration when finished.


# 7.  What are the advantages of using generators over regular functions?

  - Generators are better than regular functions because they save memory by producing values one at a time instead of creating and storing all results at once. This makes them ideal for working with large or infinite data streams. Generators also improve performance because they generate values only when needed, which is called lazy evaluation. Additionally, they can make your code easier to write and read when dealing with sequences.

  -For example, to get numbers from 1 to 1 million, a normal function would return a huge list that uses a lot of memory. But a generator like this:

   def generate_numbers():

    for i in range(1, 1000001):

        yield i

    - it will give numbers one by one, so it doesnâ€™t need to store all million numbers in memory at once.

# 9. Explain the purpose and usage of the `map()` function in Python.

  - The map() function in Python is used to apply a specific function to every item in an iterable, like a list or tuple, and returns a map object with the results. This helps in processing or transforming data efficiently without using a loop. You can convert the result into a list or other collection types if needed.

  - For example, if you have a list [1, 2, 3, 4] and want to add 10 to each number, you can use map() with a lambda function like this:

  numbers = [1, 2, 3, 4]

  result = map(lambda x: x + 10, numbers)

  print(list(result))  # Output: [11, 12, 13, 14]

  -map() applies the addition to each element and gives the updated values.




# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

  - map() applies a function to every element in an iterable and returns a new iterable with the transformed values.

  Example: Adding 5 to each number in a list:

  numbers = [1, 2, 3]

  result = map(lambda x: x + 5, numbers)

  print(list(result))  
  Output: [6, 7, 8]


  - filter() uses a function to test each element in an iterable and returns only the elements where the function returns True.

  Example: Keeping only numbers greater than 2:

  numbers = [1, 2, 3, 4]

  result = filter(lambda x: x > 2, numbers)

  print(list(result))  
  
   Output: [3, 4]

  - reduce() combines all elements in an iterable into a single value by applying a function repeatedly.

  Example: Summing all numbers in a list:

  from functools import reduce

  numbers = [1, 2, 3, 4]

  result = reduce(lambda x, y: x + y, numbers)

  print(result)  
  Output: 10

```python
# 1. Write a Python function that takes a list of numbers as input and returns
# the sum of all even numbers in the list.

def sum_even_numbers(numbers):
    sum = 0
    for num in numbers:
        if num % 2 == 0:
            sum += num
    return sum

numbers = sum_even_numbers([0,1,2,4,8,10])
print(numbers)
```

```python
# 2.  Create a Python function that accepts a string and returns the
# reverse of that string.

def reverse_string(NAME):
    return NAME[::-1]

result = reverse_string('Jyoti')
print(result)

result = reverse_string('India won')
print(result)

```

```python
# 3.  Implement a Python function that takes a list of integers and returns
# a new list containing the squares of each number.

def square_numbers(numbers):
    squares = []
    for num in numbers:
        squares.append(num ** 2)
    return squares

numbers = square_numbers([1, 2, 3, 4, 5])
print(numbers)
```

```python
# 4. Write a Python function that checks if a given number is prime or
# not from 1 to 200.

def is_prime(number):
    if number < 2:
        return False
    for a in range(2, int(number ** 0.5) + 1):
        if number % a == 0:
            return False
    return True


for number in range(1, 201):
    if is_prime(number):
        print(number)

```

```python
# 6. Write a generator function in Python that yields the powers of 2 up
# to a given exponent.

def power (exponent):
  for a in range (exponent):
    yield 2**a

p= power(8)
print(next(p))
print(next(p))
print(next(p))
print(next(p))
print(next(p))
print(next(p))
print(next(p))
print(next(p))
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the
# second element of each tuple.

list_of_tuples = [('jyoti', 25), ('rahul', 28), ('sakshi', 29), ('Solu', 21)]
sorted_list = sorted(list_of_tuples, key=lambda x: x[1])
print(sorted_list)
```

```python
# 9.  Write a Python program that uses `map()` to convert a list of
# temperatures from Celsius to Fahrenheit.

celsius = [30, 40.0, 20, 40]
fahrenheit = list(map(lambda x: (x*9/5)+32, celsius))
print(fahrenheit)
```
